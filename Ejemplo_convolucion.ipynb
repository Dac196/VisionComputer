```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
ruta = 'mario.jpg'
img = cv2.imread(ruta) #bgr
imGris = cv2.imread(ruta,0) #escala de grises
imgColor = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #TRANSFORMA BGR A RGB
```

```python
# Cargar imagen en color
image_color = cv2.imread('mario.jpg')
# Convertir imagen a escala de grises
imagE = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

image_gray = cv2.cvtColor(imagE, cv2.COLOR_RGB2GRAY)
cv2_imshow(image_gray)
```

```python
kernel = np.array([[-0.8,0,1]
                 ,[-0.8,0,1]
                 ,[-0.8,0,1]])
```

```python
conv = cv2.filter2D(img,-1,kernel) #fuente, profundidad, kernel
cv2_imshow(conv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
