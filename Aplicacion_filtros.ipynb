```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
#creamos una variable
img1=cv2.imread('Imagen1.jpg')
img2=cv2.imread('Imagen2.jpg')
img3=cv2.imread('Imagen3.jpg')
img4=cv2.imread('Imagen4.jpg')
```

```python
nuevo_ancho = 200
nuevo_alto = 200
imagen_redimensionada1 = cv2.resize(img1, (nuevo_ancho, nuevo_alto))
imagen_redimensionada2 = cv2.resize(img2, (nuevo_ancho, nuevo_alto))
imagen_redimensionada3 = cv2.resize(img3, (nuevo_ancho, nuevo_alto))
imagen_redimensionada4 = cv2.resize(img4, (nuevo_ancho, nuevo_alto))
```

```python
cv2_imshow(imagen_redimensionada1)
cv2_imshow(imagen_redimensionada2)
cv2_imshow(imagen_redimensionada3)
cv2_imshow(imagen_redimensionada4)
```

```python
imagen_gris1 = cv2.imread('Imagen1.jpg',0)
imagen_gris2 = cv2.imread('Imagen2.jpg',0)
imagen_gris3 = cv2.imread('Imagen3.jpg',0)
imagen_gris4 = cv2.imread('Imagen4.jpg',0)

imagen_redimensionada1_gris = cv2.resize(imagen_gris1, (nuevo_ancho, nuevo_alto))
imagen_redimensionada2_gris = cv2.resize(imagen_gris2, (nuevo_ancho, nuevo_alto))
imagen_redimensionada3_gris = cv2.resize(imagen_gris3, (nuevo_ancho, nuevo_alto))
imagen_redimensionada4_gris = cv2.resize(imagen_gris4, (nuevo_ancho, nuevo_alto))

# Mostrar la imagen redimensionada
cv2_imshow(imagen_redimensionada1_gris)
cv2_imshow(imagen_redimensionada2_gris)
cv2_imshow(imagen_redimensionada3_gris)
cv2_imshow(imagen_redimensionada4_gris)
```

**Filtros paso alto**

Se utiliza un filtro de enfoque, que resalta los detalles de alta frecuencia en la imagen para mejorar su nitidez para despues aplicar un filtro laplaciano y tener un mejor resultado al final

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])

filtro_enfoque1 = cv2.filter2D(imagen_redimensionada1, -1, kernel)
laplacian1 = cv2.Laplacian(filtro_enfoque1,cv2.CV_32F)
filtro_laplaciano1 = cv2.convertScaleAbs(laplacian1)

filtro_enfoque2 = cv2.filter2D(imagen_redimensionada2, -1, kernel)
laplacian2 = cv2.Laplacian(filtro_enfoque2,cv2.CV_32F)
filtro_laplaciano2 = cv2.convertScaleAbs(laplacian2)

filtro_enfoque3 = cv2.filter2D(imagen_redimensionada3, -1, kernel)
laplacian3 = cv2.Laplacian(filtro_enfoque3,cv2.CV_32F)
filtro_laplaciano3 = cv2.convertScaleAbs(laplacian3)

filtro_enfoque4 = cv2.filter2D(imagen_redimensionada4, -1, kernel)
laplacian4 = cv2.Laplacian(filtro_enfoque4,cv2.CV_32F)
filtro_laplaciano4 = cv2.convertScaleAbs(laplacian4)

cv2_imshow(filtro_laplaciano1)
cv2_imshow(filtro_laplaciano2)
cv2_imshow(filtro_laplaciano3)
cv2_imshow(filtro_laplaciano4)
```

```python
alpha = 50  # Factor de realce
filtro_paso_alto1 = cv2.addWeighted(imagen_redimensionada1_gris, alpha, cv2.GaussianBlur(imagen_redimensionada1_gris, (0, 0), 3), -alpha, 0)
filtro_paso_alto2 = cv2.addWeighted(imagen_redimensionada2_gris, alpha, cv2.GaussianBlur(imagen_redimensionada2_gris, (0, 0), 3), -alpha, 0)
filtro_paso_alto3 = cv2.addWeighted(imagen_redimensionada3_gris, alpha, cv2.GaussianBlur(imagen_redimensionada3_gris, (0, 0), 3), -alpha, 0)
filtro_paso_alto4 = cv2.addWeighted(imagen_redimensionada4_gris, alpha, cv2.GaussianBlur(imagen_redimensionada4_gris, (0, 0), 3), -alpha, 0)

cv2_imshow(filtro_paso_alto1)
cv2_imshow(filtro_paso_alto2)
cv2_imshow(filtro_paso_alto3)
cv2_imshow(filtro_paso_alto4)
```

**Filtros paso bajo**

```python
kernel_size = (55,55)
sigma_x= 3
img_gauss1 = cv2.GaussianBlur(imagen_redimensionada1,kernel_size,sigma_x)
img_gauss2 = cv2.GaussianBlur(imagen_redimensionada2,kernel_size,sigma_x)
img_gauss3 = cv2.GaussianBlur(imagen_redimensionada3,kernel_size,sigma_x)
img_gauss4 = cv2.GaussianBlur(imagen_redimensionada4,kernel_size,sigma_x)

cv2_imshow(img_gauss1)
cv2_imshow(img_gauss2)
cv2_imshow(img_gauss3)
cv2_imshow(img_gauss4)
```

```python
kernel_size = (10,10)
filtro_media1 = cv2.blur(imagen_redimensionada1, kernel_size)
filtro_media2 = cv2.blur(imagen_redimensionada2, kernel_size)
filtro_media3 = cv2.blur(imagen_redimensionada3, kernel_size)
filtro_media4 = cv2.blur(imagen_redimensionada4, kernel_size)

cv2_imshow(filtro_media1)
cv2_imshow(filtro_media2)
cv2_imshow(filtro_media3)
cv2_imshow(filtro_media4)
```

**Filtros direccionales**

```python
kernel_45 = np.array([[5, 5, -3], [5, 0, -3], [-3, -3, -3]], dtype=np.float32)

kirsh_1 = cv2.filter2D(imagen_redimensionada1_gris, -1, kernel_45)
kirsh_2 = cv2.filter2D(imagen_redimensionada2_gris, -1, kernel_45)
kirsh_3 = cv2.filter2D(imagen_redimensionada3_gris, -1, kernel_45)
kirsh_4 = cv2.filter2D(imagen_redimensionada4_gris, -1, kernel_45)

cv2_imshow(kirsh_1)
cv2_imshow(kirsh_2)
cv2_imshow(kirsh_3)
cv2_imshow(kirsh_4)
```

```python
filtro_enfoque_gris1 = cv2.filter2D(imagen_redimensionada1_gris, -1, kernel)
filtro_enfoque_gris2 = cv2.filter2D(imagen_redimensionada2_gris, -1, kernel)
filtro_enfoque_gris3 = cv2.filter2D(imagen_redimensionada3_gris, -1, kernel)
filtro_enfoque_gris4 = cv2.filter2D(imagen_redimensionada4_gris, -1, kernel)

roberts_y = np.array([[0, 1], [-1, 0]], dtype=np.float32)
roberts_y1 = cv2.filter2D(filtro_enfoque_gris1, -1, roberts_y)
roberts_y2 = cv2.filter2D(filtro_enfoque_gris2, -1, roberts_y)
roberts_y3 = cv2.filter2D(filtro_enfoque_gris3, -1, roberts_y)
roberts_y4 = cv2.filter2D(filtro_enfoque_gris4, -1, roberts_y)

cv2_imshow(roberts_y1)
cv2_imshow(roberts_y2)
cv2_imshow(roberts_y3)
cv2_imshow(roberts_y4)
```

Filtros deteccin de bordes

```python
edges1 = cv2.Canny(imagen_redimensionada1, threshold1=50, threshold2=50)
edges2 = cv2.Canny(imagen_redimensionada2, threshold1=50, threshold2=50)
edges3 = cv2.Canny(imagen_redimensionada3, threshold1=50, threshold2=50)
edges4 = cv2.Canny(imagen_redimensionada4, threshold1=50, threshold2=50)

cv2_imshow(edges1)
cv2_imshow(edges2)
cv2_imshow(edges3)
cv2_imshow(edges4)
```

```python
kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32)
prewitt_x1 = cv2.filter2D(imagen_redimensionada1, -1, kernel_x)
prewitt_x2 = cv2.filter2D(imagen_redimensionada2, -1, kernel_x)
prewitt_x3 = cv2.filter2D(imagen_redimensionada3, -1, kernel_x)
prewitt_x4 = cv2.filter2D(imagen_redimensionada4, -1, kernel_x)

cv2_imshow(prewitt_x1)
cv2_imshow(prewitt_x2)
cv2_imshow(prewitt_x3)
cv2_imshow(prewitt_x4)
```
