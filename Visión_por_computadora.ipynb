```python
#mostramos la imagen
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
cv2.rectangle(img,pt1=(20,20), pt2=(100,100),color=(0,255,0), thickness=10)
plt.imshow(img)
```

```python
img_bl_negro=cv2.imread('perro.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_bl_negro, cmap='gray')
```

```python
kernel_size = (55,55)
sigma_x= 4
img_gauss = cv2.GaussianBlur(img,kernel_size,sigma_x)
plt.imshow(img_gauss)
```

```python
img_colores = cv2.bitwise_not(img)
plt.imshow(img_colores)
```

```python
cv2.rectangle(img,pt1=(20,20), pt2=(100,100),color=(0,255,0), thickness=10)
plt.imshow(img)
```

```python
kernel_size = np.ones((3, 3), np.uint8)
img_erosion = cv2.erode(img, kernel_size, iterations=3)
plt.imshow(img_erosion)
```

```python
img_dilation = cv2.dilate(img, kernel_size, iterations=3)
plt.imshow(img_dilation)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplacian)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,kernel_size)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,kernel_size)
plt.subplot(2,2,1),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
```

```python
sobel = cv2.Sobel(img,cv2.CV_64F,1,1, kernel_size)
plt.imshow(sobel)
```
